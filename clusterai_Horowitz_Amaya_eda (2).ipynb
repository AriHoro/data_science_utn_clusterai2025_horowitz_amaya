```python
# ============================================================
# EDA COMPLETO - Bank Subscription (SIN DICCIONARIO)
# Ciencia de Datos - UTN FRBA
# ============================================================

# ------------------------------------------------------------
# 0) IMPORTS Y CONFIGURACIÓN INICIAL
# ------------------------------------------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")
sns.set_palette("deep")
pd.set_option("display.max_columns", 200)

# ------------------------------------------------------------
# 1) CARGA DE DATOS
# ------------------------------------------------------------
# Ajustá el nombre del archivo si es distinto en tu Colab
DATA_PATH = "bank_subscription.csv"   # <--- cambiar si hace falta

# IMPORTANTE: el CSV viene separado por ';'
df = pd.read_csv(DATA_PATH, sep=";")
print("Shape del dataset:", df.shape)

# Vista rápida
display(df.head())

# Info general
print("\n=== INFO ===")
print(df.info())

print("\n=== DESCRIBE NUMÉRICAS ===")
display(df.describe().T)

# ------------------------------------------------------------
# 2) DEFINIR VARIABLE OBJETIVO Y TIPO DE VARIABLES
# ------------------------------------------------------------
# En este dataset la variable objetivo es 'Subscription'
TARGET_COL = "Subscription"

if TARGET_COL not in df.columns:
    print(f"\n⚠ La columna objetivo '{TARGET_COL}' no existe en el dataset. Revisar nombre.")
else:
    print(f"\n✅ Variable objetivo detectada: {TARGET_COL}")

# Separar numéricas y categóricas (sin la target)
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = df.select_dtypes(include=["object", "category", "bool"]).columns.tolist()

if TARGET_COL in numeric_cols:
    numeric_cols.remove(TARGET_COL)
if TARGET_COL in cat_cols:
    cat_cols.remove(TARGET_COL)

print("\nColumnas numéricas:", numeric_cols)
print("\nColumnas categóricas:", cat_cols)

# ------------------------------------------------------------
# 3) CALIDAD DE DATOS: FALTANTES, DUPLICADOS, CONSTANTES
# ------------------------------------------------------------

print("\n=== VALORES FALTANTES POR COLUMNA ===")
missing = df.isna().sum().sort_values(ascending=False)
display(missing.to_frame("n_missing"))

print("\n=== PORCENTAJE DE FALTANTES ===")
missing_pct = (df.isna().mean().sort_values(ascending=False) * 100).round(2)
display(missing_pct.to_frame("%missing"))

# Duplicados
n_dups = df.duplicated().sum()
print(f"\nRegistros duplicados (filas idénticas): {n_dups}")

# Columnas constantes / casi constantes
nunique = df.nunique().sort_values()
print("\n=== CANTIDAD DE VALORES ÚNICOS POR COLUMNA ===")
display(nunique.to_frame("n_unique"))

const_cols = nunique[nunique == 1].index.tolist()
print("\nColumnas constantes (un solo valor):", const_cols)

# ------------------------------------------------------------
# 4) ANÁLISIS DE LA VARIABLE OBJETIVO
# ------------------------------------------------------------
if TARGET_COL in df.columns:
    print("\n=== DISTRIBUCIÓN DE LA VARIABLE OBJETIVO ===")
    display(df[TARGET_COL].value_counts())
    display((df[TARGET_COL].value_counts(normalize=True) * 100).round(2))

    plt.figure(figsize=(4,4))
    df[TARGET_COL].value_counts().plot(kind="bar")
    plt.title("Distribución de la variable objetivo")
    plt.xlabel("Clase")
    plt.ylabel("Cantidad")
    plt.tight_layout()
    plt.show()

    # Métrica de desbalance
    vc = df[TARGET_COL].value_counts(normalize=True)
    if len(vc) == 2:
        minor = vc.min()
        major = vc.max()
        ratio = round(minor / major, 3)   # <--- CORREGIDO

        print(f"\nRatio clase minoritaria / mayoritaria: {ratio}")
        if ratio < 0.5:
            print("⚠ Dataset desbalanceado: conviene usar métricas robustas (AUC, recall, precision) y/o class_weight / re-muestreo.")
        else:
            print("✅ Dataset moderadamente balanceado.")
    else:
        print("\nLa variable objetivo tiene más de 2 clases (multiclase).")

# ------------------------------------------------------------
# 5) DISTRIBUCIONES DE VARIABLES NUMÉRICAS
# ------------------------------------------------------------

print("\n=== HISTOGRAMAS DE VARIABLES NUMÉRICAS ===")
for col in numeric_cols:
    plt.figure(figsize=(5,4))
    sns.histplot(df[col].dropna(), kde=True)
    plt.title(f"Histograma de {col}")
    plt.tight_layout()
    plt.show()

    desc = df[col].describe(percentiles=[0.01,0.25,0.5,0.75,0.99]).T
    print(f"\nResumen estadístico de {col}:")
    display(desc.to_frame().T)

# ------------------------------------------------------------
# 6) BOXPLOTS + OUTLIERS EN NUMÉRICAS
# ------------------------------------------------------------

print("\n=== BOXPLOTS Y DETECCIÓN SIMPLE DE OUTLIERS (IQR) ===")
outlier_report = []

for col in numeric_cols:
    data = df[col].dropna()
    if len(data) == 0:
        continue

    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    n_outliers = ((data < lower) | (data > upper)).sum()
    pct_outliers = (n_outliers / len(data) * 100).round(2)

    outlier_report.append({
        "variable": col,
        "Q1": Q1,
        "Q3": Q3,
        "IQR": IQR,
        "lower_whisker": lower,
        "upper_whisker": upper,
        "n_outliers": n_outliers,
        "%_outliers": pct_outliers,
    })

    plt.figure(figsize=(4,4))
    sns.boxplot(x=data)
    plt.title(f"Boxplot de {col}")
    plt.tight_layout()
    plt.show()

if len(outlier_report) > 0:
    outlier_df = pd.DataFrame(outlier_report).sort_values("%_outliers", ascending=False)
    print("\n== Resumen de outliers por variable numérica (ordenado por % de outliers) ==")
    display(outlier_df)
else:
    print("\nNo hay variables numéricas para analizar outliers.")

# ------------------------------------------------------------
# 7) VARIABLES CATEGÓRICAS: FRECUENCIAS
# ------------------------------------------------------------

print("\n=== DISTRIBUCIÓN DE VARIABLES CATEGÓRICAS ===")
for col in cat_cols:
    print(f"\n>>> {col}")
    display(df[col].value_counts().head(20))        # top 20 categorías
    display((df[col].value_counts(normalize=True).head(20) * 100).round(2))

    plt.figure(figsize=(6,4))
    df[col].value_counts().head(10).plot(kind="bar")
    plt.title(f"{col} - Top 10 categorías")
    plt.xlabel(col)
    plt.ylabel("Cantidad")
    plt.tight_layout()
    plt.show()

# ------------------------------------------------------------
# 8) RELACIÓN TARGET vs VARIABLES NUMÉRICAS
# ------------------------------------------------------------

if TARGET_COL in df.columns and len(numeric_cols) > 0:
    print("\n=== TARGET vs VARIABLES NUMÉRICAS ===")
    for col in numeric_cols:
        plt.figure(figsize=(6,4))
        sns.boxplot(x=TARGET_COL, y=col, data=df)
        plt.title(f"{col} vs {TARGET_COL}")
        plt.tight_layout()
        plt.show()

        plt.figure(figsize=(6,4))
        sns.kdeplot(data=df, x=col, hue=TARGET_COL, common_norm=False)
        plt.title(f"Densidad de {col} por {TARGET_COL}")
        plt.tight_layout()
        plt.show()

# ------------------------------------------------------------
# 9) RELACIÓN TARGET vs VARIABLES CATEGÓRICAS
# ------------------------------------------------------------

if TARGET_COL in df.columns:
    print("\n=== TARGET vs VARIABLES CATEGÓRICAS ===")
    for col in cat_cols:
        print(f"\n>>> {col}")
        crosstab = pd.crosstab(df[col], df[TARGET_COL])
        prop = pd.crosstab(df[col], df[TARGET_COL], normalize="index")
        print("Tabla de contingencia:")
        display(crosstab)
        print("Proporción por categoría:")
        display((prop * 100).round(2))

        # Barplot de proporción de la clase "positiva" si es binaria
        if df[TARGET_COL].nunique() == 2:
            positive_class = df[TARGET_COL].value_counts().index[0]  # clase más frecuente
            plt.figure(figsize=(7,4))
            prop_pos = prop[positive_class].sort_values(ascending=False)
            prop_pos.head(20).plot(kind="bar")
            plt.title(f"Proporción de {positive_class} por {col} (top 20 categorías)")
            plt.ylabel("Proporción")
            plt.tight_layout()
            plt.show()

# ------------------------------------------------------------
# 10) MATRIZ DE CORRELACIÓN ENTRE NUMÉRICAS
# ------------------------------------------------------------

if len(numeric_cols) > 1:
    print("\n=== MATRIZ DE CORRELACIÓN (NUMÉRICAS) ===")
    corr_cols = numeric_cols.copy()
    # si la target es numérica, la incluimos
    if TARGET_COL in df.columns and df[TARGET_COL].dtype != "O":
        corr_cols = corr_cols + [TARGET_COL]

    corr = df[corr_cols].corr()
    display(corr)

    plt.figure(figsize=(10,8))
    sns.heatmap(corr, annot=False, cmap="coolwarm", center=0)
    plt.title("Heatmap de correlación")
    plt.tight_layout()
    plt.show()

    # Ver pares con correlación alta
    corr_pairs = (
        corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
        .stack()
        .reset_index()
    )
    corr_pairs.columns = ["var1", "var2", "correlation"]
    high_corr = corr_pairs[abs(corr_pairs["correlation"]) >= 0.7].sort_values("correlation", ascending=False)
    print("\nPares de variables con |correlación| >= 0.7:")
    display(high_corr)
else:
    print("\nNo hay suficientes variables numéricas para matriz de correlación.")

# ------------------------------------------------------------
# 11) ENCODING SUGERIDO PARA CATEGÓRICAS (sólo setup)
# ------------------------------------------------------------

print("\n=== SUGERENCIA DE ENCODING PARA VARIABLES CATEGÓRICAS ===")
encoding_plan = []
for col in cat_cols:
    n_cat = df[col].nunique()
    encoding_plan.append({
        "variable": col,
        "n_categorias": n_cat,
        "sugerencia": "One-Hot (dummies)" if n_cat < 20 else "One-Hot con rare categories / agrupación previa"
    })

encoding_plan_df = pd.DataFrame(encoding_plan).sort_values("n_categorias", ascending=False)
display(encoding_plan_df)

# ------------------------------------------------------------
# 12) RESUMEN Y PRÓXIMOS PASOS PARA EL TRABAJO FINAL
# ------------------------------------------------------------

print("\n============================================================")
print("RESUMEN EDA - GUÍA PARA EL RESTO DEL TRABAJO")
```

# 1. Descripción del problema y del conjunto de datos

El banco dispone de información de campañas de marketing telefónico realizadas a **45.211** clientes.  
Cada registro corresponde a un contacto con un cliente potencial e incluye:

- Características sociodemográficas: **Age**, **Job**, **Marital Status**, **Education**.  
- Información financiera: **Balance (euros)**, **Credit**, **Housing Loan**, **Personal Loan**.  
- Información de la campaña actual: **Contact**, **Last Contact Day**, **Last Contact Month**, **Last Contact Duration**, **Campaign**.  
- Historial de campañas anteriores: **Pdays**, **Previous**, **Poutcome**.  

La variable objetivo es **Subscription**, que indica si el cliente **contrató (1)** o **no contrató (0)** el producto ofrecido.  
Se trata de un problema de **clasificación binaria supervisada**.

La distribución de la variable objetivo es fuertemente desbalanceada:

- **Subscription = 0:** 88,3 %  
- **Subscription = 1:** 11,7 %

---

# 2. Principales hallazgos del análisis exploratorio (EDA)

## 2.1. Calidad de datos

- Existen faltantes de magnitud intermedia:
  - ~17 % en: **Personal Loan**, **Housing Loan**, **Pdays**, **Last Contact Duration**.  
  - ~11 % en: **Age**, **Job**, **Marital Status**, **Education**, **Credit**, **Balance (euros)**.
- No se detectaron filas duplicadas.  
- No hay variables constantes.

Esto obliga a definir una estrategia clara de **imputación de datos faltantes** y a tratar con cuidado las variables ligadas al historial de campañas, en especial **Pdays** (donde el valor -1 tiene significado propio).

---

## 2.2. Variables numéricas

- **Age**
  - Media ≈ 41 años, concentrada entre 30 y 50.  
  - Algunos valores elevados aparecen como *outliers*.

- **Balance (euros)**
  - Distribución altamente asimétrica.  
  - Muchos clientes con saldo bajo y pocos con saldos muy altos (y también negativos).

- **Last Contact Day**
  - Distribución bastante uniforme entre 1 y 31.

- **Last Contact Duration**
  - Distribución muy asimétrica.  
  - La mayoría de las llamadas son cortas; existen llamadas muy largas (*outliers*).

- **Campaign**
  - La mayoría de los clientes recibe entre 1 y 3 contactos.  
  - Cola larga de casos con muchos contactos (outliers).

- **Pdays** y **Previous**
  - `Pdays = -1` y `Previous = 0` en la mayor parte de los registros.  
  - Sólo una fracción menor de clientes tiene historial de campañas anteriores.

**Correlaciones**

- No hay pares de variables numéricas con \|correlación\| ≥ 0,7.  
- Destaca la correlación positiva entre **Last Contact Duration** y **Subscription** (~0,40), lo que sugiere que la duración de la llamada es un predictor relevante del éxito de la campaña.

---

## 2.3. Variables categóricas

A partir de las frecuencias y de las tasas de suscripción por categoría se observan estos patrones:

- **Job**
  - Segmentos como *student* y *retired* presentan tasas de suscripción sensiblemente mayores al promedio.  
  - *Management* y *unemployed* también responden mejor que *blue-collar* o *services*.

- **Marital Status**
  - Los *single* tienen una probabilidad de suscripción mayor que los *married*.

- **Education**
  - La probabilidad de suscripción crece con el nivel educativo:  
    **tertiary > secondary > primary**.

- **Credit**, **Housing Loan**, **Personal Loan**
  - Los clientes con créditos vigentes tienden a suscribirse menos que aquellos sin créditos.

- **Contact**
  - El canal **cellular** es el más utilizado y tiene mayor tasa de suscripción que **unknown** o **telephone**.

- **Last Contact Month**
  - Meses con menor volumen de llamadas (por ejemplo *mar*, *oct*, *sep*, *dec*) presentan tasas de suscripción muy altas.  
  - *May*, que concentra el mayor número de contactos, muestra una tasa de conversión baja.

- **Poutcome** (resultado de la campaña anterior)
  - Cuando el resultado previo fue **success**, la tasa de suscripción actual ronda el 65 %.  
  - Cuando el resultado es **unknown** (mayoría de los casos), la tasa cae a alrededor de 9 %.

---

## 2.4. Relación conjunta con la suscripción

De los boxplots y curvas de densidad `variable vs Subscription` se desprende que:

- Los clientes que se suscriben tienden a:
  - tener llamadas más largas (**Last Contact Duration**),  
  - poseer saldos algo mayores (**Balance (euros)**),  
  - y, en promedio, ser ligeramente mayores en edad (**Age**).
- A medida que aumenta el número de contactos en la campaña actual (**Campaign**), la probabilidad de suscripción tiende a disminuir, sugiriendo un posible efecto de saturación.
- La combinación de **Pdays**, **Previous** y **Poutcome** confirma que:
  - la base está compuesta principalmente por **primeros contactos** (sin historial),  
  - y que los clientes con campañas previas exitosas constituyen un grupo pequeño pero muy propenso a suscribirse.

---

## 2.5. Insight central del EDA

En síntesis, el EDA sugiere que:

1. La campaña se dirige principalmente a clientes **sin historial previo de contacto** (`Pdays = -1`, `Previous = 0`).  
2. Hay diferencias claras entre:
   - clientes sin historial,  
   - clientes con historial no exitoso,  
   - y clientes con campañas anteriores exitosas (`Poutcome = success`).  
3. Las variables del contacto actual (**Last Contact Duration**, **Campaign**, **Last Contact Month**, **Contact**) y del perfil del cliente (**Job**, **Education**, **Marital Status**) son fuertemente discriminantes de la respuesta.

Este diagnóstico motiva, además del modelo global, el análisis por **segmentos** (sin historial vs con historial), utilizando el mismo tipo de modelo.

---

# 3. Estrategia de preprocesamiento

## 3.1. Tratamiento de datos faltantes

- **Variables numéricas**
  - Imputación por mediana en: **Age**, **Balance (euros)**, **Last Contact Duration**.  
  - En **Pdays**, se mantiene explícito el valor `-1` para indicar “sin contacto previo”; los demás faltantes se imputan por mediana si fuera necesario.

- **Variables categóricas**
  - Se agrega una categoría explícita `Missing` en:  
    **Job**, **Marital Status**, **Education**, **Credit**, **Housing Loan**, **Personal Loan**.

---

## 3.2. Variables derivadas

- **has_previous_contact**
  - 1 si el cliente tiene algún contacto previo (`Pdays > -1` o `Previous > 0` o `Poutcome ≠ 'unknown'`).  
  - 0 si no hay historial.

- **poutcome_success**
  - 1 si `Poutcome = 'success'`.  
  - 0 en cualquier otro caso.

---

## 3.3. Tratamiento de outliers

Para reducir el impacto de valores extremos sin perder información se propone:

- Winsorización o truncamiento en percentiles **P1–P99** de:
  - **Balance (euros)**  
  - **Campaign**  
  - **Previous**  
  - **Last Contact Duration**

---

## 3.4. Codificación de variables categóricas

Se aplica **One-Hot Encoding** a:

- **Job**, **Marital Status**, **Education**, **Credit**,  
  **Housing Loan**, **Personal Loan**, **Contact**,  
  **Last Contact Month**, **Poutcome**,  
  y eventualmente a las variables derivadas binarias.

---

## 3.5. Manejo del desbalance de clases

En el entrenamiento se utilizará:

- `class_weight = "balanced"` en el modelo, para penalizar más los errores sobre la clase minoritaria (`Subscription = 1`) y mejorar el **recall** de suscriptores.

---

# 4. Estrategia de modelado: comparación de múltiples algoritmos

## 4.1. Motivación general

Si bien el árbol de decisión es un modelo interpretativo y muy adecuado para explicar reglas de negocio, el EDA mostró patrones complejos, no lineales, interacciones entre variables y un fuerte desbalance de clases.  
Por este motivo, resulta conveniente **no limitarse a un único algoritmo**, sino comparar el desempeño de **todos los modelos vistos en la cursada**, y evaluar cuál se adapta mejor a:

- el **dataset completo**,  
- el **segmento sin historial de contacto**,  
- el **segmento con historial previo no exitoso**,  
- el **segmento con historial exitoso**.

Con esta estrategia se busca:
- obtener un modelo global robusto,  
- entender cómo cambian los patrones según el tipo de cliente,  
- y seleccionar el algoritmo que maximiza el desempeño en cada segmento.

---

## 4.2. Modelos incluidos (todos vistos en la cursada)

Los modelos a entrenar serán:

- **Regresión Logística**  
- **K-Nearest Neighbors (KNN)**  
- **Árbol de Decisión**  
- **Random Forest**  
- **Gradient Boosting** (si estuviera disponible en mljar-supervised)  
- **Support Vector Machine (SVM)**  

Todos estos algoritmos fueron trabajados en clase y permiten evaluar diferentes propiedades:

- modelos **lineales vs no lineales**,  
- modelos **de bordes suaves vs reglas discretas**,  
- modelos de **bajo sesgo** vs **baja varianza**,  
- modelos sensibles a la **escala** vs modelos que no requieren escalado,  
- modelos robustos al **ruido, outliers y variables categóricas**.

---

## 4.3. Uso de MLJAR para automatizar la comparación

Se utilizará **MLJAR-Supervised** para:

- correr todos los modelos de forma estandarizada,  
- realizar búsqueda automática de hiperparámetros,  
- obtener métricas comparables (AUC, F1, Recall, Confusion Matrix),  
- y generar reportes y gráficos automáticos.

MLJAR permite replicar en minutos un procedimiento equivalente a lo que haríamos manualmente con GridSearchCV o RandomizedSearchCV, pero con documentación integrada y resultados visuales.

# 5. Reducción de dimensionalidad con PCA

## 5.1. Motivación

Tras el One-Hot Encoding, el dataset tendrá muchas columnas.  
Incluso con la eliminación de categorías raras, la dimensionalidad crecerá significativamente.

Por ello se realizará un segundo ciclo de modelado aplicando **PCA (Principal Component Analysis)** a las variables numéricas y dummy, con el objetivo de:

- reducir el ruido,  
- capturar la mayor varianza explicada con menos componentes,  
- y mejorar el rendimiento de modelos sensibles a la dimensionalidad (por ejemplo, KNN y SVM).

## 5.2. Procedimiento

1. Estandarizar variables numéricas y columnas dummy.  
2. Aplicar PCA conservando, por ejemplo, un **80–90 %** de la varianza explicada.  
3. Usar estas nuevas componentes como input para los mismos modelos del apartado anterior.  
4. Comparar si la reducción dimensional mejora o empeora rendimiento y estabilidad.

Esto permitirá obtener conclusiones explicando:

- qué modelos se benefician del PCA,  
- en qué segmentos,  
- y si existe un "sweet spot" entre dimensionalidad y performance.
